```python
from json import loads

import matplotlib.pyplot as plt
import numpy as np
```

```python
def matches(row, q):
    if isinstance(q, list):
        return isinstance(row, list) and all(matches(subrow, subq) for subrow, subq in zip(subrow, q))
    elif isinstance(q, dict):
        return isinstance(row, dict) and all(key in row and matches(row[key], q[key]) for key in q.keys())
    elif isinstance(q, float):
        return isinstance(row, float) and abs(q - row) < 1e-6
    else:
        return row == q

def query(q):
    res = []
    with open("db.txt", "r") as fd:
        for i, line in enumerate(fd):
            data = loads(line)
            if matches(data, q):
                res.append({**data, "_id": i+1})
    return res

def list_query(q):
    res = []
    for data in query(q):
        del data["payload"]["losses"]
        del data["_id"]
        del data["date"]
        res.append(data)
    return res

def list_db():
    return list_query({})
```

```python
plt.figure(figsize=(28, 14))
for c, lr in zip("krbgy", [1e-2, 2e-2, 3e-2, 4e-2, 5e-2]):
    for data in query({"payload": {"kwargs": {"init_lr": lr, "random_seed": 3}}}):
        if "epochs" in data["payload"]["kwargs"] or "N_f" in data["payload"]["kwargs"]:
            continue
        seed = data["payload"]["kwargs"]["random_seed"]
        losses = np.log10(data["payload"]["losses"])
        plt.plot(losses, f"-{c}", label=f"lr={lr}")
plt.legend()
plt.title("Different learning rates")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```

```python
for epochs in [1000, 2000, 3000, 4000, 5000, 7000, 9000, 10000]:
    plt.figure(figsize=(28, 6))
    for c, N_f in zip("rgbcmyk", [5000, 10000, 20000, 40000, 80000, 100000]):
        for data in query({"payload": {"kwargs": {"init_lr": 0.02, "random_seed": 3, "N_f": N_f, "epochs": epochs}}}):
            if "meta" in data["payload"]:
                continue
            losses = np.log10(data["payload"]["losses"][500:])
            plt.plot(losses, c=c, label=f"N_f={N_f}")
    plt.legend()
    plt.title("Different training points count")
    plt.xlabel("epoch")
    plt.ylabel("loss")
    plt.show()
```

```python
plt.figure(figsize=(28, 14))
ode = r"""$$\frac{0.01}{\pi}u_{xx} - u_t - uu_x = 0, x ∈ [−1, 1], t ∈ [0, 1]$$
$$\begin{cases}
u(0, x) = − sin(\pi x) \\
u(t, −1) = u(t, 1) = 0
\end{cases}$$"""
for data in query({"payload": {"ode": ode}}):
    if not "random_seed" in data["payload"]["kwargs"]:
        continue
    OFFSET = 500
    losses = np.log10(data["payload"]["losses"][OFFSET:])
    N_f = data["payload"]["kwargs"]["N_f"]
    random_seed = data["payload"]["kwargs"]["random_seed"]
    s = "rgbyk"[random_seed - 1] + {100:"-", 500:"-.", 1000:"--", 2000:"--", 3000:".", 4000:"."}[N_f]
    plt.plot(np.arange(len(losses)) + OFFSET, losses, s, label=f"N_f={N_f}, random_seed={random_seed}")
plt.legend()
plt.title("Different N_f")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```
